# drop_duplicates()

Return DataFrame with duplicate rows removed. Considering certain columns is optional. Indexes, including time indexes are ignored.

Syntax: DataFrame.drop_duplicates(subset=None, keep=â€™firstâ€™, inplace=False)

Parameters:

1.  subset: Subset takes a column or list of column label. Itâ€™s default value is none. After passing columns, it will consider them only for duplicates.
2.  keep: keep is to control how to consider duplicate value. It has only three distinct value and default is â€˜firstâ€™.
    * If â€˜firstâ€˜, it considers first value as unique and rest of the same values as duplicate.
    * If â€˜lastâ€˜, it considers last value as unique and rest of the same values as duplicate.
    * If False, it consider all of the same values as duplicates
3.  inplace: Boolean values, removes rows with duplicates if True. Makes the permenant changes implace
* Return type: DataFrame with removed duplicate rows depending on Arguments passed.

```python
data.drop_duplicates(subset = ['city'])
```

```python
data.drop_duplicates(subset = ['city']).shape
```

drop_duplicates() on multiple columns
* works as an AND operation

```python
data.drop_duplicates(subset = ['city', 'season'])
```

```python
data.drop_duplicates(subset = ['city', 'season']).shape
```

```python
# display every season winner

data.drop_duplicates('season', keep='last')[['season', 'winner']].sort_values('season')
```

# **isin()**

**Find the most destructive death over batsman in the history of IPL**

strike Rate = (Number of runs / number of balls)*100

Min batsman 200 balls in over 16-20

```python
mask = delivery['over'] > 15

death_over = delivery[mask]
```

```python
all_batsman = death_over.groupby('batsman')['batsman_runs'].count()

x = all_batsman > 200

batsman_list = all_batsman[x].index.tolist()
```

To Calculate strike rate
1.  Runs scored by all these 43 batsman
2.  Balls played by these 43 batsman

```python
mask2 = death_over['batsman'].isin(batsman_list)

final = death_over[mask2]
```

```python
runs = final.groupby('batsman')['batsman_runs'].sum()
```

```python
balls = final.groupby('batsman')['batsman_runs'].count()
```

```python
sr = (runs/balls)*100
```

```python
sr
```

# **corr function**

Pandas dataframe.corr() is used to find the pairwise correlation of all columns in the Pandas Dataframe
* Syntax: DataFrame.corr(self, method=â€™pearsonâ€™, min_periods=1, numeric_only = False)

Parameters:

* method :
    * pearson: standard correlation coefficient
    * kendall: Kendall Tau correlation coefficient
    * spearman: Spearman rank correlation
* min_periods : Minimum number of observations required per pair of columns to have a valid result. Currently only available for pearson and spearman correlation
* numeric_only : Whether only the numeric values are to be operated upon or not. It is set to False by default.

```python
matches.head()
```

```python
# returns a pivot table and only work for numeric value

matches.corr(numeric_only = True)
```

```python
sns.heatmap(matches.corr(numeric_only = True))
```
